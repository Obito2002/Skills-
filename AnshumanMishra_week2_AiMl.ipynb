As we can see here the the missing values are of the columns club , value, International Reputation , Skill Moves and contract valid until.
We can use the mean of the columns International Reputation,Value, Skill Moves to replace the null values.
As for the column Club and contract valid until , there is no resonable values that can be used to replace the null values.


```python
raw_data.rename(columns = {'Preferred Foot':'P_Foot', 'International Reputation':'I_Reputation','Release Clause':'R_clause','Contract Valid Until':'cvu','Skill Moves' : 'S_move' }, inplace = True)
```

```python
raw_data.head()
```

```python
raw_data.drop(['cvu'],axis=1)
```

```python
raw_data['I_Reputation'] = raw_data['I_Reputation'].replace(np.nan , 1.113222 )
raw_data['Value'] = raw_data['Value'].replace(np.nan , 2444.530214 )
raw_data['S_move'] = raw_data['S_move'].replace(np.nan , 2.361308 )
```

```python
raw_data.isnull().sum()
```

Now that we have replaced the null values in the data, we shal now drop the rest of the null values.

```python
new=raw_data.dropna()
```

```python
new.isnull().sum()
```

```python
new.describe()
```

Here we can see no major changes occur on the data as a whole after deleting the null values.
